### Data Extraction

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#BY TOTAL ASSETS
def extract_table1(url):
#url =  'https://en.wikipedia.org/wiki/List_of_largest_banks'
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    table1 = soup.find('table')
    header = [headers.text.strip() for headers in table1.find_all('th')]
    df = pd.DataFrame(columns= header)
# rows = table.find_all('tr')
    for row in table1.find_all('tr')[1:]:
        row.find('td')
        individual_row = [data.text.strip() for data in row]
        filtered_data = [item for item in individual_row if item != '']
        length = len(df)
        df.loc[length] = filtered_data
    return df
        #print(filtered_data)
def df_to_csv(banks_dataframe):
   banks_dataframe = banks_dataframe.to_csv('G:\Data engineering\Web Scrapping\Bank_data (by total assets).csv', index=False)
   return banks_dataframe
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_largest_banks' 

def extract_table2(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    table2 = soup.find_all('table')[1]
    rows = []
    for row in table2.find_all('tr'):
        cells = [cell.text.strip() for cell in row.find_all('td')]
        if len(cells) == 2 or len(cells) == 3:
            rows.append(cells)
    normalized_data = []
    for row in rows:
        if len(row) == 3:
            normalized_data.append(row)
        elif len(row) == 2:
            normalized_data.append(['', row[0], row[1]]) 
    df = pd.DataFrame(normalized_data, columns=['Rank', 'Country', 'Number'])    
    return df
data = extract_table2(url)
print(data)
```

```python
def df_into_csv(data):
    data = data.to_csv('G:\Data engineering\Web Scrapping\Bank_data (By country number).csv', index=False)
    return data
```
