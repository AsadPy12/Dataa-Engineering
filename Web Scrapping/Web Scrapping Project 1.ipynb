```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
print(soup)
```

```python
table = soup.find('table', class_= 'wikitable sortable')
```

```python
title = table.find_all('th')
```

```python
heading = [header.text.strip() for header in title]
```

```python
heading
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = heading)
df
```

```python
data = table.find_all('tr')
data
```

```python
for row in data[1:]:
    row_data = row.find_all('td')
    indivisual_row_data = [rows.text.strip() for rows in row_data]
    #print(indivisual_row_data)
    
    length = len(df)
    df.loc[length] = indivisual_row_data
```

```python
df
```

```python
df.to_csv('G:\Data engineering\Web Scrapping\CompaniesData.csv', index = False)
```
